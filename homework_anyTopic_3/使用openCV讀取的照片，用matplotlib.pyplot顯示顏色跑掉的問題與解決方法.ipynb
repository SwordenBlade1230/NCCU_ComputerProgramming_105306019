# 使用openCV讀取的照片，用matplotlib.pyplot顯示顏色跑掉的問題與解決方法
圖片來源：https://www.pexels.com/zh-tw/photo/1170986/

```python
### 讀取函式庫 ###
import cv2
import matplotlib.pyplot as plt
```

```python
### 讀取相片並印出相片的維度 ###
img = cv2.imread('cat.jpg')
img.shape
```

```python
plt.imshow(img) # 印出相片
```

直接用plt.imshow會發現顏色和原來的照片不一樣  
原因：openCV的色彩接口是GBR，但matplotlib.pyplot是RGB  
解決方法：把GBR編排成RGB的形式

```python
b,g,r = cv2.split(img)
img_rgb = cv2.merge([r,g,b])
plt.imshow(img_rgb) # 印出相片
```
